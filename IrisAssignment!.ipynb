```python
import numpy as np
import pandas as pd
import seaborn as snbrn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the Data set
columns = ["Sepal_length","Sepal_width","petal_Length","petal_width","Class_labels"]
#Load the data
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=columns)
df.head()
```

```python
#Visualization of our dataset
df.describe()
```

```python
snbrn.pairplot(df,hue='Class_labels')
```

```python
#Separate features and target
data = df.values
X = data[:,0:4]
Y = data[:,4]
```

```python
#Splitting the data into train and test samples
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
#Model 1:Support Vector machine
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1 = model_svc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1))
```

```python
#Model 2 : Logistic Regression
from sklearn.linear_model import LogisticRegression
model_Log_reg = LogisticRegression()
model_Log_reg.fit(X_train,Y_train)
prediction2 = model_Log_reg.predict(X_test)
#calcualte accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2)*100)
```
